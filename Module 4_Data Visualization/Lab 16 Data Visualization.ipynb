**Read the CSV File into a Pandas DataFrame**

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("survey-data.sqlite")

# Read the CompTotal column
QUERY = """
SELECT CompTotal
FROM main
WHERE CompTotal IS NOT NULL
"""

df_comp = pd.read_sql_query(QUERY, conn)

# Close the connection
conn.close()

# Plot histogram
plt.figure(figsize=(10,6))
plt.hist(df_comp["CompTotal"], bins=30, edgecolor="black")

plt.title("Histogram of Total Compensation (CompTotal)")
plt.xlabel("CompTotal")
plt.ylabel("Frequency")

plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
## Write your code here

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("survey-data.sqlite")

# Read Age and WorkExp columns
QUERY = """
SELECT Age, WorkExp
FROM main
WHERE Age IS NOT NULL
  AND WorkExp IS NOT NULL
"""

df_scatter = pd.read_sql_query(QUERY, conn)

# Close the connection
conn.close()

# Convert Age categories to numeric values
age_mapping = {
    "Under 18 years old": 17,
    "18-24 years old": 21,
    "25-34 years old": 29,
    "35-44 years old": 39,
    "45-54 years old": 49,
    "55-64 years old": 59,
    "65 years or older": 70
}

df_scatter["Age"] = df_scatter["Age"].map(age_mapping)

# Plot scatter plot
plt.figure(figsize=(8,6))
plt.scatter(df_scatter["Age"], df_scatter["WorkExp"], alpha=0.5)

plt.title("Scatter Plot of Age vs WorkExp")
plt.xlabel("Age")
plt.ylabel("Work Experience")

plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
## Write your code here

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("survey-data.sqlite")

# Read required columns
QUERY = """
SELECT Age, TimeSearching, Frustration
FROM main
WHERE Age IS NOT NULL
  AND TimeSearching IS NOT NULL
  AND Frustration IS NOT NULL
"""

df_bubble = pd.read_sql_query(QUERY, conn)

# Close the connection
conn.close()

# Convert Age categories to numeric values
age_mapping = {
    "Under 18 years old": 17,
    "18-24 years old": 21,
    "25-34 years old": 29,
    "35-44 years old": 39,
    "45-54 years old": 49,
    "55-64 years old": 59,
    "65 years or older": 70
}

df_bubble["Age"] = df_bubble["Age"].map(age_mapping)

# Convert categorical columns to numeric codes
df_bubble["TimeSearching"] = df_bubble["TimeSearching"].astype("category").cat.codes
df_bubble["Frustration"] = df_bubble["Frustration"].astype("category").cat.codes

# Create bubble plot
plt.figure(figsize=(10,6))

plt.scatter(
    df_bubble["TimeSearching"],
    df_bubble["Frustration"],
    s=df_bubble["Age"] * 5,
    alpha=0.5
)

plt.title("Bubble Plot of TimeSearching vs Frustration")
plt.xlabel("TimeSearching")
plt.ylabel("Frustration")

plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
## Write your code here

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("survey-data.sqlite")

# Read the DatabaseWantToWorkWith column
QUERY = """
SELECT DatabaseWantToWorkWith
FROM main
WHERE DatabaseWantToWorkWith IS NOT NULL
"""

df_db = pd.read_sql_query(QUERY, conn)

# Close the connection
conn.close()

# Split multiple databases into separate rows
db_counts = (
    df_db["DatabaseWantToWorkWith"]
    .str.split(";")
    .explode()
    .value_counts()
    .head(5)
)

# Plot pie chart
plt.figure(figsize=(8,8))

plt.pie(
    db_counts,
    labels=db_counts.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Top 5 Databases Respondents Want to Learn Next Year")
plt.axis("equal")

plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
## Write your code here

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("survey-data.sqlite")

# Read required columns
QUERY = """
SELECT Age, TimeSearching, TimeAnswering
FROM main
WHERE Age IS NOT NULL
  AND TimeSearching IS NOT NULL
  AND TimeAnswering IS NOT NULL
"""

df_stack = pd.read_sql_query(QUERY, conn)

# Close the connection
conn.close()

# Map Age to numeric values
age_mapping = {
    "Under 18 years old": 17,
    "18-24 years old": 21,
    "25-34 years old": 29,
    "35-44 years old": 39,
    "45-54 years old": 49,
    "55-64 years old": 59,
    "65 years or older": 70
}

df_stack["Age"] = df_stack["Age"].map(age_mapping)

# Filter age group 30–35
df_stack = df_stack[(df_stack["Age"] >= 30) & (df_stack["Age"] <= 35)]

# Convert categorical columns to numeric codes
df_stack["TimeSearching"] = df_stack["TimeSearching"].astype("category").cat.codes
df_stack["TimeAnswering"] = df_stack["TimeAnswering"].astype("category").cat.codes

# Calculate median values
median_search = df_stack["TimeSearching"].median()
median_answer = df_stack["TimeAnswering"].median()

# Plot stacked bar chart
plt.figure(figsize=(6,6))

plt.bar("Age 30-35", median_search, label="TimeSearching")
plt.bar("Age 30-35", median_answer, bottom=median_search, label="TimeAnswering")

plt.title("Median TimeSearching and TimeAnswering (Age 30-35)")
plt.ylabel("Median Value")
plt.legend()

plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("survey-data.sqlite")

# Read Age and CompTotal columns
QUERY = """
SELECT Age, CompTotal
FROM main
WHERE Age IS NOT NULL
  AND CompTotal IS NOT NULL
"""

df_line = pd.read_sql_query(QUERY, conn)

# Close the connection
conn.close()

# Map Age categories to numeric values
age_mapping = {
    "Under 18 years old": 17,
    "18-24 years old": 21,
    "25-34 years old": 29,
    "35-44 years old": 39,
    "45-54 years old": 49,
    "55-64 years old": 59,
    "65 years or older": 70
}

df_line["Age"] = df_line["Age"].map(age_mapping)

# Filter ages from 45 to 60
df_line = df_line[(df_line["Age"] >= 45) & (df_line["Age"] <= 60)]

# Calculate median CompTotal by Age
median_comp = df_line.groupby("Age")["CompTotal"].median()

# Plot line chart
plt.figure(figsize=(8,5))
plt.plot(median_comp.index, median_comp.values, marker='o')

plt.title("Median CompTotal for Ages 45 to 60")
plt.xlabel("Age")
plt.ylabel("Median CompTotal")

plt.grid(True)
plt.show()
```
