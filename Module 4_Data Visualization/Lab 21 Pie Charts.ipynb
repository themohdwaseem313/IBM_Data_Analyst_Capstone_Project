### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here

import pandas as pd

# Use existing dataframe assumption: df contains DatabaseWantToWorkWith
# If not loaded earlier, user can run SQL part again.

# Split multiple database values
db = df["DatabaseWantToWorkWith"].dropna().str.split(";").explode()

# Get top 5 databases
top5 = db.value_counts().head(5)

# Plot pie chart
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))
plt.pie(
    top5.values,
    labels=top5.index,
    autopct="%1.1f%%",
    startangle=140
)

plt.title("Top 5 Databases Respondents Want to Work With")
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here

import pandas as pd
import matplotlib.pyplot as plt

# Assume df already contains DevType column from dataset
# If not, load it from SQLite

import sqlite3
conn = sqlite3.connect("survey-data.sqlite")

QUERY = """
SELECT DevType
FROM main
WHERE DevType IS NOT NULL
"""

df = pd.read_sql_query(QUERY, conn)
conn.close()

# Split multiple roles
dev = df["DevType"].str.split(";").explode()

# Get top 5 developer roles
top5 = dev.value_counts().head(5)

# Plot pie chart
plt.figure(figsize=(8,8))

plt.pie(
    top5.values,
    labels=top5.index,
    autopct="%1.1f%%",
    startangle=140
)

plt.title("Top 5 Developer Roles Distribution")
plt.show()
```
