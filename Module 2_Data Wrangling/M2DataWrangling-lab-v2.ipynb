### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

#### 2. Explore the Dataset


<h5>2.1 Summarize the dataset by displaying the column data types, counts, and missing values.</h5>


```python
# Write your code here

# Create a summary of the dataset
summary = pd.DataFrame({
    'Data Type': df.dtypes,
    'Non-Null Count': df.count(),
    'Missing Values': df.isnull().sum()
})

# Display the summary
print(summary)
```

<h5>2.2 Generate basic statistics for numerical columns.</h5>


```python
# Write your code here

numerical_summary = df.describe()

print(numerical_summary)
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
abs# Display unique country values
print("Unique Country Values:")
print(df['Country'].unique())

# Display frequency of each country
print("\nCountry Value Counts:")
print(df['Country'].value_counts(dropna=False))
```

```python
# Find rows where Country is missing
missing_country = df[df['Country'].isnull()]

print("Number of missing Country values:", missing_country.shape[0])

# Display first few rows with missing Country
missing_country.head()
```

```python
# Find blank Country entries
blank_country = df[df['Country'].astype(str).str.strip() == ""]

print("Number of blank Country entries:", blank_country.shape[0])

blank_country.head()
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here

# Remove leading/trailing spaces
df['Country'] = df['Country'].str.strip()

# Standardize common country names
country_mapping = {
    'USA': 'United States of America',
    'U.S.A.': 'United States of America',
    'United States': 'United States of America',
    'US': 'United States of America',
    'UK': 'United Kingdom',
    'U.K.': 'United Kingdom'
}

# Replace inconsistent values
df['Country'] = df['Country'].replace(country_mapping)

# Verify
print(df['Country'].value_counts().head(10))
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here

# One-Hot Encode the Employment column
employment_encoded = pd.get_dummies(
    df,
    columns=['Employment'],
    prefix='Employment'
)

# Display the first 5 rows
employment_encoded.head()
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here

# Count missing values in each column
missing_values = df.isnull().sum()

# Sort columns by missing values (highest first)
missing_values = missing_values.sort_values(ascending=False)

# Display columns with missing values
print(missing_values[missing_values > 0])
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here

# Check missing values before imputation
print("Missing values before:")
print(df['ConvertedCompYearly'].isnull().sum())

# Calculate median
median_value = df['ConvertedCompYearly'].median()

# Fill missing values with the median
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(median_value)

# Verify missing values after imputation
print("\nMissing values after:")
print(df['ConvertedCompYearly'].isnull().sum())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here

# Check missing values before imputation
print("Missing values before:")
print(df['RemoteWork'].isnull().sum())

# Find the most frequent (mode) value
most_frequent = df['RemoteWork'].mode()[0]

print("\nMost Frequent Value:", most_frequent)

# Fill missing values with the most frequent value
df['RemoteWork'] = df['RemoteWork'].fillna(most_frequent)

# Verify missing values after imputation
print("\nMissing values after:")
print(df['RemoteWork'].isnull().sum())
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here

# Find the minimum and maximum values
min_value = df['ConvertedCompYearly'].min()
max_value = df['ConvertedCompYearly'].max()

# Apply Min-Max Scaling
df['ConvertedCompYearly_MinMax'] = (
    (df['ConvertedCompYearly'] - min_value) /
    (max_value - min_value)
)

# Display the first few rows
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_MinMax']].head())
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
import numpy as np

# Apply log transformation
df['ConvertedCompYearly_Log'] = np.log1p(df['ConvertedCompYearly'])

# Display the first few rows
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_Log']].head())
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here

# Handle special values in YearsCodePro (if any)
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0,
    'More than 50 years': 51
})

# Convert to numeric
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Create ExperienceLevel column based on conditions

def get_experience_level(years):
    if years < 2:
        return 'Beginner'
    elif years < 5:
        return 'Intermediate'
    elif years < 10:
        return 'Advanced'
    else:
        return 'Expert'

df['ExperienceLevel'] = df['YearsCodePro'].apply(get_experience_level)

# Display result
df[['YearsCodePro', 'ExperienceLevel']].head()
```
