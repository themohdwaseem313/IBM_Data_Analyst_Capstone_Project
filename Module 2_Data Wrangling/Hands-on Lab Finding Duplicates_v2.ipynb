## Identify and Analyze Duplicates


### Task 1: Identify Duplicate Rows
1. Count the number of duplicate rows in the dataset.
3. Display the first few duplicate rows to understand their structure.


```python
## Write your code here

# 1. Task 1: Identify Duplicate Rows
Count the number of duplicate rows in the dataset.
duplicate_count = df.duplicated().sum()

print("Number of duplicate rows:", duplicate_count)

# Display the first few duplicate rows
duplicate_rows = df[df.duplicated()]

duplicate_rows.head()
```

### Task 2: Analysis of Duplicate Response Patterns
1. Identify duplicate response patterns based on selected columns such as MainBranch, Employment, and RemoteWork.
2. Clarify that these represent multiple respondents with identical answers rather than duplicate records. Analyse which other columns frequently share identical values within these response-pattern groups.
   


```python
## Write your code here
abs# Columns to check response patterns
pattern_cols = ['MainBranch', 'Employment', 'RemoteWork']

# Find response-pattern duplicates
response_patterns = df[df.duplicated(subset=pattern_cols, keep=False)]

print("Number of rows with identical response patterns:",
      response_patterns.shape[0])

# Display the first few matching response patterns
response_patterns[pattern_cols].head(10)
```

### Task 3: Visualize Shared Response Patterns
1. Create visualizations to show the distribution of shared response patterns across different categories.
2. Use bar charts or pie charts to represent the distribution of respondents who share identical values for MainBranch, Employment, and RemoteWork, grouped by Country and Employment.


```python
## Write your code here


# -----------------------------
# Identify Shared Response Patterns
# -----------------------------
pattern_cols = ['MainBranch', 'Employment', 'RemoteWork']

# Find respondents having identical response patterns
response_patterns = df[df.duplicated(subset=pattern_cols, keep=False)]

# Display total records with shared response patterns
print("Number of respondents with shared response patterns:",
      response_patterns.shape[0])

# Display first few records
print(response_patterns.head())

```

```python
# -----------------------------
# Bar Chart: Distribution by Country
# -----------------------------
country_counts = response_patterns['Country'].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.bar(country_counts.index, country_counts.values)

plt.title("Top 10 Countries with Shared Response Patterns")
plt.xlabel("Country")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# -----------------------------
# Pie Chart: Distribution by Employment
# -----------------------------
employment_counts = response_patterns['Employment'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(
    employment_counts,
    labels=employment_counts.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Employment Distribution of Shared Response Patterns")
plt.axis('equal')
plt.show()
```

```python
# -----------------------------
# Analysis of Other Shared Columns
# -----------------------------
print("\nNumber of unique values within each response-pattern group:\n")

shared_columns = response_patterns.groupby(pattern_cols).nunique()

print(shared_columns.head())
```
