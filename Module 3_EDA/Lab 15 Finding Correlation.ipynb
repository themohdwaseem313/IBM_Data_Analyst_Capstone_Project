### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
## Write your code here

import matplotlib.pyplot as plt
import seaborn as sns

# Remove missing values
comp = df['ConvertedCompYearly'].dropna()

# Create figure
plt.figure(figsize=(12,5))

# Distribution Plot (KDE)
plt.subplot(1,2,1)
sns.kdeplot(comp, fill=True)
plt.title("Distribution of ConvertedCompYearly")
plt.xlabel("ConvertedCompYearly")

# Histogram
plt.subplot(1,2,2)
plt.hist(comp, bins=30)
plt.title("Histogram of ConvertedCompYearly")
plt.xlabel("ConvertedCompYearly")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
## Write your code here

# Filter respondents who are employed full-time
full_time = df[df['Employment'] == 'Employed, full-time']

# Calculate the median yearly compensation
median_compensation = full_time['ConvertedCompYearly'].median()

# Display the result
print("Median Yearly Compensation (Employed, full-time):", median_compensation)
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
## Write your code here

import matplotlib.pyplot as plt
import seaborn as sns

# Remove missing values
comp_country = df[['Country', 'ConvertedCompYearly']].dropna()

# Select the top 10 countries by number of respondents
top_countries = comp_country['Country'].value_counts().head(10).index

# Filter data for the top 10 countries
comp_country = comp_country[comp_country['Country'].isin(top_countries)]

# Create box plot
plt.figure(figsize=(14, 6))

sns.boxplot(
    data=comp_country,
    x='Country',
    y='ConvertedCompYearly'
)

plt.title("Yearly Compensation Distribution by Country")
plt.xlabel("Country")
plt.ylabel("ConvertedCompYearly")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
## Write your code here

# Calculate Q1, Q3, and IQR for ConvertedCompYearly
Q1 = df['ConvertedCompYearly'].quantile(0.25)
Q3 = df['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create a new DataFrame without outliers
df_refined = df[
    (df['ConvertedCompYearly'] >= lower_bound) &
    (df['ConvertedCompYearly'] <= upper_bound)
]

# Validate the new DataFrame
print("Original Dataset Shape:", df.shape)
print("Refined Dataset Shape:", df_refined.shape)
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
## Write your code here

import seaborn as sns
import matplotlib.pyplot as plt

# Select required columns and remove missing values
corr_df = df_refined[['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']].dropna()

# Calculate correlation matrix
correlation_matrix = corr_df.corr(method='pearson')

# Display correlation matrix
print(correlation_matrix)

# Visualize the correlation matrix
plt.figure(figsize=(6,5))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    vmin=-1,
    vmax=1
)

plt.title("Correlation Matrix")
plt.show()
```
